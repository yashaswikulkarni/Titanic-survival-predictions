```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/Users/Yashaswi/DATA SCIENCE/TITANIC/train.csv',header=0)
data.head()
```

```python
data.shape
```

```python
from collections import Counter    # to ignore these 3 columns as they are not of any use for prediction
```

```python
data.Age.replace('Age',np.nan,inplace=True)
data.Embarked.replace([{'S': '0'},{'C':'1'},{'Q':'2'}],inplace=True)
data.drop(labels='Cabin',axis=1)
```

```python
data.Age.fillna(value=(data.Age.mean()),inplace=True)
data.Embarked.fillna('S',inplace=True)
data.Embarked.isnull().sum()
# dropping missing values with reference to age and embarked columns
```

```python
from sklearn import preprocessing
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
from sklearn.preprocessing import LabelEncoder
enc =LabelEncoder()
```

```python
data.Sex = enc.fit_transform(data.Sex)
data.Embarked = enc.fit_transform(data.Embarked)
```

```python
X = data.loc[:,['Sex','Age','Fare','Parch','Pclass','Embarked']]
y = data.Survived
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=15,stratify=y)
```

```python
model = XGBClassifier(n_estimators=100,n_jobs=10,random_state=25)
model.fit(X_train,y_train)
```

```python
 y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```
